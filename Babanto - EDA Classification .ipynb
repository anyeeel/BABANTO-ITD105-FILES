```python
# 5. Handling Missing Data - identify and handle missing values using techniques 
# like imputation or removal.

missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)

df = df.fillna(df.mean())
```

```python
# 7. Data Visualization - Create visualizations to explore data distributions, relationships, and patterns. 
#Use libraries like Matplotlib and Seaborn.

#a. Univariate plots focus on a single variable or feature at a time. 
#They are used to visualize the distribution and characteristics of individual variables in isolation.

#Histogram
plt.figure(figsize=(6, 4))
df['label'].value_counts().plot(kind='bar')
plt.xlabel('Diseases')
plt.ylabel('Count')
plt.title('Distribution of Disease Labels')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df['text_length'] = df['text'].apply(len)

plt.figure(figsize=(8, 4))
sns.kdeplot(df['text_length'], fill=True)
plt.title('Density Plot of Text Length')
plt.xlabel('Text Length')
plt.ylabel('Density')
plt.show()
```

```python
# Box and Whisker Plots
# Calculate text lengths for each symptom description
df['text_length'] = df['text'].apply(len)

# Create a box plot for text lengths grouped by disease labels
plt.figure(figsize=(12, 4))
sns.boxplot(x='label', y='text_length', data=df)
plt.xticks(rotation=90)
plt.xlabel('Disease Labels')
plt.ylabel('Text Length')
plt.title('Distribution of Text Lengths for Each Disease Label')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 'text' and 'label' columns
# Add the 'text_length' column as you've already done

# Create a box plot for text lengths grouped by disease labels
plt.figure(figsize=(12, 4))
sns.boxplot(x='label', y='text_length', data=df)
plt.xticks(rotation=90)
plt.xlabel('Disease Labels')
plt.ylabel('Text Length')
plt.title('Distribution of Text Lengths for Each Disease Label')
plt.show()

# Calculate the correlation matrix
correlation_matrix = df.corr(method='pearson')

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix (Pearson Method)')
plt.show()
```
